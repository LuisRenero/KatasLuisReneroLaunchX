### Ejercicio2

```python
planet1 = input('Ingresa la distancia del primer planeta: ')
planet2 = input('Ingresa la distancia del segundo planeta: ')
```

```python
planet1 = int(planet1)
planet2 = int(planet2)

distance = abs(planet1 - planet2)

distance_miles = distance*0.621

print(f'La distancia entre los planetas en km es: {distance} \n La distancia entre planetas en millas es: {distance_miles}')
```
