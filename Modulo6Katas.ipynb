## Ejercicio 2: Trabajando con datos de una lista

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

```python
input_planet = input('Ingresa el nombre de un planeta en el sistema solar: ')
```

```python
find_planet_index = planets.index(input_planet)
```

```python
planets_closer = planets[0:find_planet_index]
print(f'Los planetas más cercanos al sol que el ingresado son: {planets_closer}')
planets_further = planets[find_planet_index + 1:]
print(f'Los planetas que se encuentran más lejos del ingresado son: {planets_further}')
```
