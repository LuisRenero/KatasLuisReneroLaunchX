```python
def genera_reporte(destino, *minutes, **fuel_reservoirs):
    reporte = f"""
    El destino del viaje es: {destino}
    El tiempo en minutos es: {minutes}
    El combustible en los tanques es: {fuel_reservoirs}
    """
    return reporte

print(genera_reporte('Luna',10,12,15,23,tanque1=10,tanque2=100))
```

```python
def genera_reporte(destino, *minutes, **fuel_reservoirs):
    reporte = f"""
    El destino del viaje es: {destino}
    El tiempo en minutos es: {minutes}
    El combustible en los tanques es: {sum(fuel_reservoirs.values())}
    ==============================================
    """

    for item in fuel_reservoirs:
        reporte = reporte + f"""El tanque {item} tiene {fuel_reservoirs.get(item)}% de combustible \n"""
    
    return reporte

print(genera_reporte('Luna',10,12,15,23,tanque1=10,tanque2=100))
```
