# Ejercicio 1: Transformar cadenas

```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""
```

```python
text_split = text.split(". ")
print(text_split)
```

```python
keywords = ["average", "temperature", "distance"]
print(keywords)
```

```python
for text_split_item in text_split:
    for keyword_item in keywords:
        if(keyword_item in text_split_item):
            print(text_split_item)
```

```python
for text_split_item in text_split:
    for keyword_item in keywords:
        if(keyword_item in text_split_item):
            if(" C" in text_split_item):
                print(text_split_item.replace(" C", " Celsius"))
            else: print(text_split_item)
```

# Ejercicio 2: Formateando Cadenas

```python
# Datos con los que vas a trabajar
name = "Moon"
gravity = 0.00162 # in kms
planet = "Earth"
```

```python
title = f'The gravity in {planet} and its {name}'
print(f'{title.title()}')
```

```python
information_template = """
==================================
planet's name: {planet}
name of its moon: {name}
gravity: {gravity}'
==================================
"""
print(information_template.format(planet=planet, name=name, gravity=gravity*100))
```

```python
print(f'{title.title() + information_template.format(planet=planet, name=name, gravity=gravity*100)}')
```
